Este documento tiene como funcion concatenar los documentos .csv que contienen la infomacion de las doce rutas

```python
import pandas as pd
```

```python
datos_uno=pd.read_csv("../Data/NYC_PAR.csv")
datos_uno.head()
```

```python
datos_dos=pd.read_csv("../Data/NYC_RUH.csv")
datos_dos.head()
```

```python
datos_tres=pd.read_csv("../Data/NYC_SVO.csv")
datos_tres.head()
```

```python
datos_cuatro=pd.read_csv("../Data/PAR_NYC.csv")
datos_cuatro.head()
```

```python
datos_cinco=pd.read_csv("../Data/PAR_RUH.csv")
datos_cinco.head()
```

```python
datos_seis=pd.read_csv("../Data/PAR_SVO.csv")
datos_seis.head()
```

```python
datos_siete=pd.read_csv("../Data/RUH_NYC.csv")
datos_siete.head()
```

```python
datos_ocho=pd.read_csv("../Data/RUH_PAR.csv")
datos_ocho.head()
```

```python
datos_nueve=pd.read_csv("../Data/RUH_SVO.csv")
datos_nueve.head()
```

```python
datos_diez=pd.read_csv("../Data/SVO_NYC.csv")
datos_diez.head()
```

```python
datos_once=pd.read_csv("../Data/SVO_PAR.csv")
datos_once.head()
```

```python
datos_doce=pd.read_csv("../Data/SVO_RUH.csv")
datos_doce.head()
```

```python
Datos_completos=pd.concat([datos_uno,datos_dos,datos_tres,datos_cuatro,datos_cinco,
                           datos_seis,datos_siete,datos_ocho,datos_nueve,datos_diez,
                           datos_once,datos_doce], ignore_index=True)
```

```python
Datos_completos
```

```python
Datos_completos.to_csv("../Data/Datos_completos.csv",index=False)
```
